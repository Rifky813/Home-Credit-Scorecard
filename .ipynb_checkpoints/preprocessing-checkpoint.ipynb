### Impute Numerical, Binary, and Categorical Missing Values

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
```

```python
def fit_imputer(df):

    # Deteksi kolom numerik & binary
    binary_cols = [col for col in df.select_dtypes(include=['int64', 'float64']).columns
                   if set(df[col].dropna().unique()) <= {0, 1}]
    
    num_cols = [col for col in df.select_dtypes(include=['int64', 'float64']).columns
                    if col not in binary_cols]

    cat_cols = ['NAME_TYPE_SUITE']
    
    imputer_mode = SimpleImputer(strategy='most_frequent')
    imputer_med = SimpleImputer(strategy='median')
    
    imputer = ColumnTransformer(
        transformers=[
            ('bin', imputer_mode, binary_cols), 
            ('num', imputer_med, num_cols),
            ('cat', imputer_mode, cat_cols)
        ],
        remainder='passthrough'
    )

    imputer.fit(df)

    return imputer, binary_cols + num_cols + cat_cols
```

```python
def transform_imputer(df, imputer, ordered_cols):
    df_array = imputer.transform(df)

    # Ambil nama semua kolom (yang ikut maupun tidak ikut transformasi)
    all_columns = ordered_cols + [col for col in df.columns if col not in ordered_cols]

    # Kembalikan dataframe hasil imputasi
    return pd.DataFrame(df_array, columns=all_columns, index=df.index)
```

```python
original_dtypes = apptr.dtypes

X = apptr.drop('TARGET', axis=1)
y = apptr['TARGET']

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=50)
```

```python
imputer, imputed_cols = fit_imputer(X_train)

X_train_filled = transform_imputer(X_train, imputer, imputed_cols)
X_test_filled  = transform_imputer(X_test, imputer, imputed_cols)
appts_filled  = transform_imputer(appts, imputer, imputed_cols)
```

```python
X_train_filled.shape
```

```python
def ori_dtype(df):
    for col in df.columns:
        if col in original_dtypes:
            try:
                df[col] = df[col].astype(original_dtypes[col])
            except Exception as e:
                print(f"Gagal casting kolom '{col}' ke {original_dtypes[col]}: {e}")
    return df
```

```python
X_train_filled = ori_dtype(X_train_filled)
X_test_filled = ori_dtype(X_test_filled)
appts_filled = ori_dtype(appts_filled)
```

### Encoding Numerical Variable

```python
from sklearn.preprocessing import LabelEncoder

le_dict = {}
le_count = 0

for col in X_train_filled.select_dtypes(include='object').columns:
    if X_train_filled[col].nunique() == 2:
        le = LabelEncoder()
        le.fit(X_train_filled[col])
        
        X_train_filled[col] = le.transform(X_train_filled[col])
        X_test_filled[col] = X_test_filled[col].map(lambda x: le.transform([x])[0] if x in le.classes_ else np.nan)
        appts_filled[col] = appts[col].map(lambda x: le.transform([x])[0] if x in le.classes_ else np.nan)
        
        le_dict[col] = le
        le_count += 1

print(f"{le_count} columns were label encoded")
```

### Encoding Categorical Variable

```python
# Tentukan kolom kategorikal
cat_cols = X_train_filled.select_dtypes(include='object').columns.tolist()

# Pisahkan berdasarkan jumlah kategori
low_card_cols = [col for col in cat_cols if X_train_filled[col].nunique() <= 10]
high_card_cols = [col for col in cat_cols if X_train_filled[col].nunique() > 10]

# One-Hot Encoding untuk low cardinality
X_train_low = pd.get_dummies(X_train_filled[low_card_cols], drop_first=True)
X_test_low = pd.get_dummies(X_test_filled[low_card_cols], drop_first=True)
appts_low = pd.get_dummies(appts_filled[low_card_cols], drop_first=True)

# Pastikan kolom sama di train & test (karena drop_first bisa bikin beda)
X_test_low = X_test_low.reindex(columns=X_train_low.columns, fill_value=0)
appts_low = appts_low.reindex(columns=X_train_low.columns, fill_value=0)

# Encode high-cardinality dengan reduksi top-N kategori
top_k = 15
X_train_high_encoded = []
X_test_high_encoded = []
appts_high_encoded = []

for col in high_card_cols:
    top_cats = X_train_filled[col].value_counts().nlargest(top_k).index
    
    # Replace rare dengan 'Other'
    train_reduced = X_train_filled[col].where(X_train_filled[col].isin(top_cats), 'Other')
    test_reduced = X_test_filled[col].where(X_test_filled[col].isin(top_cats), 'Other')
    appts_reduced = appts_filled[col].where(appts_filled[col].isin(top_cats), 'Other')
    
    # One-hot encode
    train_encoded = pd.get_dummies(train_reduced, prefix=col, drop_first=True)
    test_encoded = pd.get_dummies(test_reduced, prefix=col, drop_first=True)
    appts_encoded = pd.get_dummies(appts_reduced, prefix=col, drop_first=True)
    
    # Reindex test supaya sama kolomnya
    test_encoded = test_encoded.reindex(columns=train_encoded.columns, fill_value=0)
    appts_encoded = appts_encoded.reindex(columns=train_encoded.columns, fill_value=0)
    
    X_train_high_encoded.append(train_encoded)
    X_test_high_encoded.append(test_encoded)
    appts_high_encoded.append(appts_encoded)

# Gabungkan semua hasil encoding
X_train_encoded = pd.concat(
    [X_train_filled.drop(columns=cat_cols), X_train_low] + X_train_high_encoded, axis=1)

X_test_encoded = pd.concat(
    [X_test_filled.drop(columns=cat_cols), X_test_low] + X_test_high_encoded, axis=1)

appts_encoded = pd.concat(
    [appts_filled.drop(columns=cat_cols), appts_low] + appts_high_encoded, axis=1)

# Cek hasil akhir
print("Train encoded shape:", X_train_encoded.shape)
print("Test encoded shape:", X_test_encoded.shape)
print("application_test encoded shape:", appts_encoded.shape)
```

### SelectKBest Method

#### Using chi2
Scale the numeric variable first using MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler

col_name = X_train_encoded.columns

mms = MinMaxScaler()
X_train_scaled = mms.fit_transform(X_train_encoded)
X_test_scaled = mms.transform(X_test_encoded)
final_appts = mms.transform(appts_encoded)
```

```python
from sklearn.feature_selection import SelectKBest, chi2

bestfeatures = SelectKBest(score_func=chi2, k=30)
fit_chi2 = bestfeatures.fit(X_train_scaled, y_train)

dfscores = pd.DataFrame(fit_chi2.scores_)
dfcolumns = pd.DataFrame(col_name)
featureScores = pd.concat([dfcolumns, dfscores], axis = 1)
featureScores.columns = ['Features', 'Score']

pd.set_option('display.max_rows', None)
print('The features that correlate well with target feature:\n')
featureScores.sort_values(by=['Score'], ascending=False)
```

```python
X_train_selected_chi2 = fit_chi2.transform(X_train_scaled)
X_test_selected_chi2 = fit_chi2.transform(X_test_scaled)
appts_selected_chi2 = fit_chi2.transform(final_appts)

columns_chi2 = X_train_encoded.columns[fit_chi2.get_support()]

X_train_chi2 = pd.DataFrame(X_train_selected_chi2, columns=columns_chi2)
X_test_chi2 = pd.DataFrame(X_test_selected_chi2, columns=columns_chi2)
final_appts_chi2 = pd.DataFrame(appts_selected_chi2, columns=columns_chi2)
```

#### Using Mutual Info

```python
from sklearn.feature_selection import mutual_info_classif

bestfeatures_mi = SelectKBest(score_func=mutual_info_classif, k=30)
fit_mi = bestfeatures_mi.fit(X_train_encoded, y_train)

dfscores = pd.DataFrame(fit_mi.scores_)
dfcolumns = pd.DataFrame(col_name)
featureScores = pd.concat([dfcolumns, dfscores], axis = 1)
featureScores.columns = ['Features', 'Score']

pd.set_option('display.max_rows', None)
print('The features that correlate well with target feature:\n')
featureScores.sort_values(by=['Score'], ascending=False)
```

```python
X_train_selected_mi = fit_mi.transform(X_train_encoded)
X_test_selected_mi = fit_mi.transform(X_test_encoded)
appts_selected_mi = fit_mi.transform(final_appts)

columns_mi = X_train_encoded.columns[fit_mi.get_support()]

X_train_mi = pd.DataFrame(X_train_selected_mi, columns=columns_mi)
X_test_mi = pd.DataFrame(X_test_selected_mi, columns=columns_mi)
final_appts_mi = pd.DataFrame(appts_selected_mi, columns=columns_mi)
```
