## Examine Missing Values

```python
def check_missing_values(df):
    mv_count = df.isna().sum()
    mv_percentage = ((mv_count / len(df)) * 100).round(1).astype(str) + "%"
    df_mv = pd.concat([mv_count, mv_percentage], axis=1)

    # rename columns
    df_mv = df_mv.rename(columns={0: "Missing Value Count", 1: "Percentage of Missing Value by Total"})
    
    df_mv = df_mv.sort_values("Percentage of Missing Value by Total", ascending=False)
    
    return df_mv

missing_values = check_missing_values(app_train)
missing_values.head(20)
```

### Exterior Sources

--> Skor penilaian risiko kegagalan membayar dari lembaga eksternal, dipakai untuk bantu menilai apakah seseorang aman atau berisiko untuk diberi pinjaman.
Skor dimulai dari 0 sampai 1:
Semakin mendekati 0 artinya semakin berisiko gagal bayar
Semakin mendekati 1 artinya semakin aman atau rendah risiko

## Feature Engineering

Kita bisa membuat beberapa fitur yang mencoba menangkap hal-hal yang menurut kita penting untuk menentukan apakah klien akan gagal bayar pinjaman. Di sini saya akan menggunakan empat fitur yang terinspirasi oleh skrip Aguiar:

- CREDIT_INCOME_PERCENT: persentase total kredit relatif terhadap pendapatan klien 
- ANNUITY_INCOME_PERCENT: Persentase cicilan bulanan (annuity) terhadap pendapatan klien
- CREDIT_TERM: Lama periode cicilan dalam bulan
- DAYS_EMPLOYED_PERCENT: Persentase lama bekerja dibandingkan umur klien (dalam hari)

```python
app_train_newfeats = app_train.copy()
app_test_newfeats = app_test.copy()

app_train_newfeats['CREDIT_INCOME_PERCENT'] = app_train_newfeats['AMT_CREDIT'] / app_train_newfeats['AMT_INCOME_TOTAL']
app_train_newfeats['ANNUITY_INCOME_PERCENT'] = app_train_newfeats['AMT_ANNUITY'] / app_train_newfeats['AMT_INCOME_TOTAL']
app_train_newfeats['CREDIT_TERM'] = app_train_newfeats['AMT_ANNUITY'] / app_train_newfeats['AMT_CREDIT']
app_train_newfeats['DAYS_EMPLOYED_PERCENT'] = app_train_newfeats['DAYS_EMPLOYED'] / app_train_newfeats['DAYS_BIRTH']
```

```python
app_test_newfeats['CREDIT_INCOME_PERCENT'] = app_test_newfeats['AMT_CREDIT'] / app_test_newfeats['AMT_INCOME_TOTAL']
app_test_newfeats['ANNUITY_INCOME_PERCENT'] = app_test_newfeats['AMT_ANNUITY'] / app_test_newfeats['AMT_INCOME_TOTAL']
app_test_newfeats['CREDIT_TERM'] = app_test_newfeats['AMT_ANNUITY'] / app_test_newfeats['AMT_CREDIT']
app_test_newfeats['DAYS_EMPLOYED_PERCENT'] = app_test_newfeats['DAYS_EMPLOYED'] / app_test_newfeats['DAYS_BIRTH']
```

### Visualize the new features

```python
sns.set(style='whitegrid', font_scale=1.2)
plt.figure(figsize=(10, 15))

for i, col in enumerate(app_train_newfeats.iloc[:, -4:]):
    plt.subplot(4, 1, i+1)
    sns.kdeplot(data=app_train_newfeats, x=col, hue='TARGET', common_norm=False, linewidth=2)

    plt.title(f'Distribution of {col}')

plt.tight_layout()
```
